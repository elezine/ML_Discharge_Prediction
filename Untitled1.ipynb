```python
import numpy as np
import pandas as pd
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense,LSTM
import datetime as dt
plt.rcParams['figure.dpi'] = 300
```

```python
#Training Data:

train = pd.read_csv('indin_train.csv')

era5_tp = train['era5_noon_tp'].values
era5_tp_1daylag = np.zeros(len(era5_tp))
era5_tp_1daylag[:] = np.nan
era5_tp_1daylag[1:,] = era5_tp[:-1]
train['era5_tp_1daylag'] = era5_tp_1daylag

era5_tp_2daylag = np.zeros(len(era5_tp))
era5_tp_2daylag[:] = np.nan
era5_tp_2daylag[2:,] = era5_tp[:-2]
train['era5_tp_2daylag'] = era5_tp_2daylag

era5_tp_3daylag = np.zeros(len(era5_tp))
era5_tp_3daylag[:] = np.nan
era5_tp_3daylag[3:,] = era5_tp[:-3]
train['era5_tp_3daylag'] = era5_tp_3daylag

train = train[train['flow'].notna()]
train = train[train['era5_tp_1daylag'].notna()]
train = train[train['era5_tp_2daylag'].notna()]
train = train[train['era5_tp_3daylag'].notna()]

train = train.sort_values(by="datetime")
train['doy'] = [dt.datetime.strptime(i, '%Y-%m-%d').timetuple().tm_yday for i in train.datetime]

#Validation Data:

valid = pd.read_csv('indin_valid.csv')

era5_tp = valid['era5_noon_tp'].values
era5_tp_1daylag = np.zeros(len(era5_tp))
era5_tp_1daylag[:] = np.nan
era5_tp_1daylag[1:,] = era5_tp[:-1]
valid['era5_tp_1daylag'] = era5_tp_1daylag

era5_tp_2daylag = np.zeros(len(era5_tp))
era5_tp_2daylag[:] = np.nan
era5_tp_2daylag[2:,] = era5_tp[:-2]
valid['era5_tp_2daylag'] = era5_tp_2daylag

era5_tp_3daylag = np.zeros(len(era5_tp))
era5_tp_3daylag[:] = np.nan
era5_tp_3daylag[3:,] = era5_tp[:-3]
valid['era5_tp_3daylag'] = era5_tp_3daylag

valid = valid[valid['flow'].notna()]
valid = valid[valid['era5_tp_1daylag'].notna()]
valid = valid[valid['era5_tp_2daylag'].notna()]
valid = valid[valid['era5_tp_3daylag'].notna()]

valid = valid.sort_values(by="datetime")
valid['doy'] = [dt.datetime.strptime(i, '%Y-%m-%d').timetuple().tm_yday for i in valid.datetime]

# Test Data:

test = pd.read_csv('indin_test.csv')

era5_tp = test['era5_noon_tp'].values
era5_tp_1daylag = np.zeros(len(era5_tp))
era5_tp_1daylag[:] = np.nan
era5_tp_1daylag[1:,] = era5_tp[:-1]
test['era5_tp_1daylag'] = era5_tp_1daylag

era5_tp_2daylag = np.zeros(len(era5_tp))
era5_tp_2daylag[:] = np.nan
era5_tp_2daylag[2:,] = era5_tp[:-2]
test['era5_tp_2daylag'] = era5_tp_2daylag

era5_tp_3daylag = np.zeros(len(era5_tp))
era5_tp_3daylag[:] = np.nan
era5_tp_3daylag[3:,] = era5_tp[:-3]
test['era5_tp_3daylag'] = era5_tp_3daylag

test = test[test['flow'].notna()]
test = test[test['era5_tp_1daylag'].notna()]
test = test[test['era5_tp_2daylag'].notna()]
test = test[test['era5_tp_3daylag'].notna()]

test = test.sort_values(by="datetime")
test['doy'] = [dt.datetime.strptime(i, '%Y-%m-%d').timetuple().tm_yday for i in test.datetime]


# Holdout Data:

holdout = pd.read_csv('indin_holdout.csv')

era5_tp = holdout['era5_noon_tp'].values
era5_tp_1daylag = np.zeros(len(era5_tp))
era5_tp_1daylag[:] = np.nan
era5_tp_1daylag[1:,] = era5_tp[:-1]
holdout['era5_tp_1daylag'] = era5_tp_1daylag

era5_tp_2daylag = np.zeros(len(era5_tp))
era5_tp_2daylag[:] = np.nan
era5_tp_2daylag[2:,] = era5_tp[:-2]
holdout['era5_tp_2daylag'] = era5_tp_2daylag

era5_tp_3daylag = np.zeros(len(era5_tp))
era5_tp_3daylag[:] = np.nan
era5_tp_3daylag[3:,] = era5_tp[:-3]
holdout['era5_tp_3daylag'] = era5_tp_3daylag

holdout = holdout[holdout['flow'].notna()]
holdout = holdout[holdout['era5_tp_1daylag'].notna()]
holdout = holdout[holdout['era5_tp_2daylag'].notna()]
holdout = holdout[holdout['era5_tp_3daylag'].notna()]

holdout = holdout.sort_values(by="datetime")
holdout['doy'] = [dt.datetime.strptime(i, '%Y-%m-%d').timetuple().tm_yday for i in holdout.datetime]
```

```python
y_train = np.asarray(train.flow.values)
x_train_era  = pd.concat([train.doy, train.era5_tp_1daylag, train.era5_tp_2daylag, train.era5_tp_3daylag, train.era5_noon_t2m, train.era5_noon_evavt, train.era5_noon_ro, train.era5_noon_sd, train.era5_noon_es, train.era5_noon_sf, train.era5_noon_smlt, train.era5_noon_stl1, train.era5_noon_ssr, train.era5_noon_e, train.era5_noon_tp], axis=1)

y_test = np.asarray(test.flow.values)
x_test_era = pd.concat([test.doy, test.era5_tp_1daylag, test.era5_tp_2daylag, test.era5_tp_3daylag, test.era5_noon_t2m, test.era5_noon_evavt, test.era5_noon_ro, test.era5_noon_sd, test.era5_noon_es, test.era5_noon_sf, test.era5_noon_smlt, test.era5_noon_stl1, test.era5_noon_ssr, test.era5_noon_e, test.era5_noon_tp], axis=1)

y_valid = np.asarray(valid.flow.values)
x_valid_era = pd.concat([valid.doy, valid.era5_tp_1daylag, valid.era5_tp_2daylag, valid.era5_tp_3daylag, valid.era5_noon_t2m, valid.era5_noon_evavt, valid.era5_noon_ro, valid.era5_noon_sd, valid.era5_noon_es, valid.era5_noon_sf, valid.era5_noon_smlt, valid.era5_noon_stl1, valid.era5_noon_ssr, valid.era5_noon_e, valid.era5_noon_tp], axis=1)

y_holdout = np.asarray(holdout.flow.values)
x_holdout_era = pd.concat([holdout.doy, holdout.era5_tp_1daylag, holdout.era5_tp_2daylag, holdout.era5_tp_3daylag, holdout.era5_noon_t2m, holdout.era5_noon_evavt, holdout.era5_noon_ro, holdout.era5_noon_sd, holdout.era5_noon_es, holdout.era5_noon_sf, holdout.era5_noon_smlt, holdout.era5_noon_stl1, holdout.era5_noon_ssr, holdout.era5_noon_e, holdout.era5_noon_tp], axis=1)
```

```python
scaler = MinMaxScaler()  
x = pd.concat([x_train_era, x_test_era, x_valid_era, x_holdout_era])
x_scaled = scaler.fit(x).transform(x)

x_train = x_scaled[0:len(x_train_era)]
x_test = x_scaled[len(x_train_era):len(x_train_era)+len(x_test_era)]
x_valid = x_scaled[len(x_train_era)+len(x_test_era):len(x_train_era)+len(x_test_era)+len(x_valid_era)]
x_holdout = x_scaled[len(x_train_era)+len(x_test_era)+len(x_valid_era):]
```

---

MLP

```python
classifier = MLPRegressor(solver='adam', alpha=0.01, hidden_layer_sizes=(1000,), random_state=1, max_iter=100000)
```

```python
classifier.fit(x_train, y_train)
```

```python
predictions_test_mlp = classifier.predict(x_test)
predictions_holdout_mlp = classifier.predict(x_holdout)
predictions_valid_mlp = classifier.predict(x_valid)
```
